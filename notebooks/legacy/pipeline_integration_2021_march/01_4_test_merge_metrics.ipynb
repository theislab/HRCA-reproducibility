```python

"""
Merge metrics output for all scenarios, methods and settings
"""

parser = argparse.ArgumentParser(description='Collect all metrics')

parser.add_argument('-i', '--input', nargs='+', required=True, help='input directory')
parser.add_argument('-o', '--output', required=True, help='output file')
parser.add_argument('-r', '--root', required=True, 
                    help='root directory for inferring column names from path')
```

```python
args = parser.parse_args(args=cmd.split(' '))
# args
```

```python



scenarios = set(file.split('/')[-5] for file in args.input)
method_and_parms = set('/'.join(file.split('/')[-5:-1]) + '/' + '_'.join(file.replace(args.root, "").replace(".csv", "").split('_')[:-1]) for file in args.input)
print(method_and_parms)
print(scenarios)

res_list = []
print('\n\n\nstart')

assert len(scenarios) == 1
for scenario in scenarios:

    print('scenario')
    for next_method_and_parms in method_and_parms:

        res_scenario = []
        print('next method')
        print(next_method_and_parms.split('/')[-1])

        for file in args.input:
            curr_scenario = file.split('/')[-5]

            clean_name = '_'.join(file.replace(args.root, "").replace(".csv", "").split('_')[:-1])
            curr_method_and_parms = '/'.join(file.split('/')[-5:-1]) + '/' + clean_name

            if not curr_scenario in scenarios:
                continue

            if curr_method_and_parms != next_method_and_parms:
                continue
            if not os.path.exists(file):
                continue

            # if not '_pcr.csv' in file:
            #     continue

            # print('clean name', clean_name)
            # assert False
            # print('\nreading file', file)
            res = None
            print(curr_method_and_parms == next_method_and_parms, curr_method_and_parms, next_method_and_parms)
            print('next file', file)
            try:
                print('reading', file)
                res = pd.read_csv(file, index_col=0)
                res.rename(columns={res.columns[0]: clean_name}, inplace=True)
                res_scenario.append(res)

                # print(file)
                # print(res)
                # print('\n')
            except Exception:
                res_scenario.append(pd.DataFrame())

        print('')
            # print(res)
        print('')

        print('\n\n\n\n\nstatus before concatenation')
        # print(res_scenario)
        print(len(res_scenario))
        if len(res_scenario) == 0:
            print('skip...')
            continue
        
        # print(pd.concat(res_scenario, axis=1).columns)
        # print(pd.DataFrame(pd.concat(res_scenario, axis=1)))
        res = pd.DataFrame(pd.concat(res_scenario, axis=1).max(axis=1), columns=[next_method_and_parms])
        
        
        # print(res)
        
        res_list.append(res)

print('len of object beore concatenation')
print(len(res_list))
for resi in res_list:
    print(type(resi))
    print(res)

results = reduce(lambda  left,right: pd.merge(left, right, left_index=True, right_index=True), res_list)
results = results.T

results.to_csv(args.output.replace('.csv', '_%s.csv' % scenario))
results.T.to_csv(args.output.replace('.csv', '_T_%s.csv' % scenario))

pd.DataFrame().to_csv(args.output)
results.to_csv(args.output)

results.T


```
