```r
source('plotSingleTaskRNA.R');
 
for(pct in c('pct3', 'pct5', 'pct6')){

    if(pct != 'pct6')
        next
    print(pct)
    metrics = paste0(filesystem, 'data/', date, 'output/metrics_retina_all_cells_all_sn+sc_batch_donor_dataset_cell.type_2022_', pct, '.csv')

    print(c(file.exists(metrics), metrics))

    # next
    df <- read.csv(metrics, sep = ",")
    snsc <- df[grepl('_sn[+]sc_', df$X),]
    sc <- df[grepl('_sc_', df$X),]
    sn <- df[grepl('_sn_', df$X),]
    print(c(dim(snsc), dim(sc), dim(sn)))
    snsc_hvg1k <- snsc[grepl('HVG.1K', snsc$X),]
    snsc_hvg3k <- snsc[grepl('HVG.3K', snsc$X),]
    print(c(dim(snsc_hvg1k), dim(snsc_hvg3k)))

    snsc_hvg1k
    plotSingleTaskRNA(metrics_tab_lab=snsc_hvg1k, outdir=scib_outdir, suffix=paste0('_HVG1K_',  pct))
    snsc_hvg3k
    plotSingleTaskRNA(metrics_tab_lab=snsc_hvg3k, outdir=scib_outdir, suffix=paste0('_HVG3K_', pct))

    for(key_other in c('sn', 'sc')){
        # 1K
        other = NULL
        if(key_other == 'sn')
            other = sn
        if(key_other == 'sc')
            other = sc
        snsc_vs_other = rbind(snsc, other)

        snsc_vs_other_hvg1k <- snsc_vs_other[grepl('HVG.1K', snsc_vs_other$X),]
        dataset_group <- sapply(str_split(snsc_vs_other_hvg1k$X, "_"), function(x) x[5])
        snsc_vs_other_hvg1k$X <- str_replace(snsc_vs_other_hvg1k$X, '_sn[+]sc_', '_groups_')
        snsc_vs_other_hvg1k$X <- str_replace(snsc_vs_other_hvg1k$X, paste0('_', key_other, '_'), '_groups_')

        plotSingleTaskRNA(metrics_tab_lab=snsc_vs_other_hvg1k,
                        outdir=scib_outdir,
                        suffix=paste0('_HVG1K_', pct, '_snsc_vs_', key_other),
                        dataset_group=dataset_group)

        mask = grepl('sn[+]sc', dataset_group)
        plotSingleTaskRNA(metrics_tab_lab=snsc_vs_other_hvg1k[mask,],
                        outdir=scib_outdir,
                        suffix=paste0('_HVG1K_', pct, '_snsc_vs_', key_other, '_only_snsc'),
                        dataset_group=dataset_group[mask])

        mask = grepl(key_other, dataset_group) & !grepl('sn[+]sc', dataset_group)
        plotSingleTaskRNA(metrics_tab_lab=snsc_vs_other_hvg1k[mask,],
                        outdir=scib_outdir,
                        suffix=paste0('_HVG1K_', pct, '_snsc_vs_', key_other, '_only_', key_other),
                        dataset_group=dataset_group[mask])


        # 3K
        snsc_vs_other = rbind(snsc, other)
        snsc_vs_other_hvg3k <- snsc_vs_other[grepl('HVG.3K', snsc_vs_other$X),]
        dataset_group <- sapply(str_split(snsc_vs_other_hvg3k$X, "_"), function(x) x[5])
        dataset_group
        snsc_vs_other_hvg3k$X <- str_replace(snsc_vs_other_hvg3k$X, '_sn[+]sc_', '_groups_')
        snsc_vs_other_hvg3k$X <- str_replace(snsc_vs_other_hvg3k$X, paste0('_', key_other, '_'), '_groups_')
        plotSingleTaskRNA(metrics_tab_lab=snsc_vs_other_hvg3k,
                        outdir=scib_outdir,
                        suffix=paste0('_HVG3K_', pct, '_snsc_vs_', key_other),
                        dataset_group=dataset_group)

        plotSingleTaskRNA(metrics_tab_lab=snsc_vs_other_hvg3k,
                        outdir=scib_outdir,
                        suffix=paste0('_HVG3K_', pct, '_snsc_vs_', key_other),
                        dataset_group=dataset_group)

        mask = grepl('sn[+]sc', dataset_group)
        plotSingleTaskRNA(metrics_tab_lab=snsc_vs_other_hvg3k[mask,],
                        outdir=scib_outdir,
                        suffix=paste0('_HVG3K_', pct, '_snsc_vs_', key_other, '_only_snsc'),
                        dataset_group=dataset_group[mask])

        mask = grepl(key_other, dataset_group) & !grepl('sn[+]sc', dataset_group)
        plotSingleTaskRNA(metrics_tab_lab=snsc_vs_other_hvg3k[mask,],
                        outdir=scib_outdir,
                        suffix=paste0('_HVG3K_', pct, '_snsc_vs_sn_only_', key_other),
                        dataset_group=dataset_group[mask])
    }

    break
}
```
