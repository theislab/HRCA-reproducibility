```python
datadir = '/storage/groups/ml01/datasets/projects/20210318_retinal_data_integration_ignacio.ibarra_malte.luecken'
```

```python
import scanpy as sc
import os
sc.pl.set_rcParams_defaults()
```

```python
for f in os.listdir(os.path.join(datadir, 'Chen_a')):
    path_old = os.path.join(datadir, 'Chen_a', f)
    path_new = os.path.join(datadir, 'Chen_a_fix5', f)
    if os.path.exists(path_new):
        print(path_old)
        print(path_new)
        ad_old = sc.read_h5ad(path_old)
        ad_new = sc.read_h5ad(path_new)
        
        for ad, label in zip([ad_old, ad_new], ['old', 'new']):
            print(ad.shape)
            sc.pp.neighbors(ad)
            sc.tl.umap(ad)
            sc.pl.umap(ad, color='scpred_prediction', title=label + ', ' + f)
        print('')
```
