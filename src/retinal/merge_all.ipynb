```python
import scanpy as sc
from os import listdir
from os.path import isdir, join
```

```python
ad_final = None
n = 0
stopat = 20
for d in listdir('./'):
    if not isdir(d):
        continue
    for f in listdir(join(d)):
        p = join(d, f)
        if not p.endswith('h5ad'):
            continue
        print(p)
        adata = sc.read(p)
        print(adata.shape)
        ad_final = adata if ad_final is None else ad_final.concatenate(adata, index_unique='')
        n += 1
        if n >= stopat :
            break
    if n == stopat:
        break
print(ad_final.shape)
```

```python
ad_final.obs
```

```python
ad_final.layers['counts'] = ad_final.X
```

```python
ad_final.write('human_retinal.h5ad')
```
